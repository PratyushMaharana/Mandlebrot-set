# **Burning Ship Fractal**

```python
import numpy as np
import matplotlib.pyplot as plt

def burning_ship(c, max_iter):
    z = 0
    for n in range(max_iter):
        if abs(z) > 2:
            return n
        z = (abs(z.real) + 1j * abs(z.imag)) ** 2 + c
    return max_iter

def generate_burning_ship(width, height, x_range, y_range, max_iter):
    x = np.linspace(x_range[0], x_range[1], width)
    y = np.linspace(y_range[0], y_range[1], height)
    fractal = np.zeros((height, width), dtype=int)

    for row in range(height):
        for col in range(width):
            c = x[col] + 1j * y[row]
            fractal[row, col] = burning_ship(c, max_iter)

    return fractal

# Parameters
width, height = 800, 800
x_range = (-2.5, 1.5)
y_range = (-2, 2)
max_iter = 100

# Generate the fractal
fractal = generate_burning_ship(width, height, x_range, y_range, max_iter)

# Plot the fractal
plt.figure(figsize=(10, 10))
plt.imshow(fractal, extent=(x_range[0], x_range[1], y_range[0], y_range[1]), cmap='hot', origin='lower')
plt.colorbar(label='Iteration count')
plt.title('Burning Ship Fractal (Evil Twin of Mandelbrot)')
plt.xlabel('Re(c)')
plt.ylabel('Im(c)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

width, height = 800, 800
x_min, x_max = -2, 2
y_min, y_max = -2, 2
max_iter = 100

image = np.zeros((height, width))

for i in range(width):
    for j in range(height):
        x0 = x_min + (i / width) * (x_max - x_min)
        y0 = y_min + (j / height) * (y_max - y_min)
        c = complex(x0, y0)
        z = 0 + 0j
        iter_count = 0

        while abs(z) <= 2 and iter_count < max_iter:
            z = complex(abs(z.real), abs(z.imag)) ** 2 + c
            iter_count += 1

        image[j, i] = iter_count

plt.imshow(image.T, cmap="magma", extent=(x_min, x_max, y_min, y_max))
plt.colorbar()
plt.title("Burning Ship Fractal")
plt.show()
```
