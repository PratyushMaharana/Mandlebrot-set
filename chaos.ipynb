```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

# Image size and limits
width, height = 800, 800
x_min, x_max = -2, 2
y_min, y_max = -2, 2
max_iter = 50

# Create an empty array for pixel values
image = np.zeros((height, width))

# Unique fractal formula
for i in range(width):
    for j in range(height):
        x0 = x_min + (i / width) * (x_max - x_min)
        y0 = y_min + (j / height) * (y_max - y_min)
        c = complex(x0, y0)
        z = c
        iter_count = 0

        while abs(z) <= 4 and iter_count < max_iter:
            # Unique transformation: rotation, power mix, and perturbation
            z = (z ** 1.5) * cmath.exp(complex(0, np.sin(abs(z.real)))) + c * cmath.exp(-abs(z.imag) * 0.1)
            iter_count += 1

        image[j, i] = iter_count

# Plot the fractal
plt.imshow(image.T, cmap="twilight_shifted", extent=(x_min, x_max, y_min, y_max))
plt.colorbar()
plt.title("Chaos Web Fractal")
plt.show()
```
